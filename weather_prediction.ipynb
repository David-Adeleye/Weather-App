```python
df.drop('month',axis=1,inplace=True)
df.drop('year',axis=1,inplace=True)
```

```python
le=LabelEncoder()
df['weather']=le.fit_transform(df['weather'])
```

```python
x = df[['temp_min', 'temp_max', 'precipitation', 'wind']]
y=df['weather']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

y_pred = nb_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred, zero_division=1)


print(f"Accuracy: {accuracy:.2f}")
```

```python
print("Confusion Matrix:")
print(conf_matrix)
print("")
print("Classification Report:")
print(classification_rep)
```

```python
# Save the model to a file
with open('naive_bayes_model.pkl', 'wb') as file:
    pickle.dump(nb_model, file)
```

```python
# Load the saved model
with open('naive_bayes_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

# Feature names (replace with your actual feature names)
feature_names = ['temp_min', 'temp_max', 'precipitation', 'wind']

# Create a new data point with feature names
new_data_point = [[7.2, 11.7, 0.8, 2.3]]
new_data_point_with_names = pd.DataFrame(new_data_point, columns=feature_names)

# Make predictions on the new data point
predicted_label = loaded_model.predict(new_data_point_with_names)

print(f'Predicted Weather Label: {predicted_label[0]}')
```
